### LeNet
The LeNet-5 architecture consists of two sets of convolutional and average pooling layers, followed by a flattening convolutional layer, then two fully-connected layers and finally a softmax classifier.

```python
from IPython.display import Image
Image(filename='./images/LeNet_Original_Image.jpg') 
```

### First Layer:

The input for LeNet-5 is a 3232 grayscale image which passes through the first convolutional layer with 6 feature maps or filters having size 55 and a stride of one. The image dimensions changes from 32x32x1 to 28x28x6.

```python
from IPython.display import Image
Image(filename='./images/c1.jpg') 
```

### Second Layer:

Then the LeNet-5 applies average pooling layer or sub-sampling layer with a filter size 22 and a stride of two. The resulting image dimensions will be reduced to 14x14x6.

```python
from IPython.display import Image
Image(filename='./images/c2.jpg') 
```

### Third Layer:

Next, there is a second convolutional layer with 16 feature maps having size 55 and a stride of 1. In this layer, only 10 out of 16 feature maps are connected to 6 feature maps of the previous layer as shown below.

```python
from IPython.display import Image
Image(filename='./images/c3.png') 
```

The main reason is to break the symmetry in the network and keeps the number of connections within reasonable bounds. Thats why the number of training parameters in this layers are 1516 instead of 2400 and similarly, the number of connections are 151600 instead of 240000.

```python
from IPython.display import Image
Image(filename='./images/c4.jpg') 
```

### Fourth Layer:

The fourth layer (S4) is again an average pooling layer with filter size 22 and a stride of 2. This layer is the same as the second layer (S2) except it has 16 feature maps so the output will be reduced to 5x5x16.

```python
from IPython.display import Image
Image(filename='./images/c5.jpg') 
```

### Fifth Layer:

The fifth layer (C5) is a fully connected convolutional layer with 120 feature maps each of size 11. Each of the 120 units in C5 is connected to all the 400 nodes (5x5x16) in the fourth layer S4.

```python
from IPython.display import Image
Image(filename='./images/c6.jpg') 
```

### Sixth Layer:

The sixth layer is a fully connected layer (F6) with 84 units.

```python
from IPython.display import Image
Image(filename='./images/c7.jpg') 
```

### Output Layer:

Finally, there is a fully connected softmax output layer  with 10 possible values corresponding to the digits from 0 to 9.
